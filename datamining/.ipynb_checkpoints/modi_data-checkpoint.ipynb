```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("./kor_abs_full.xlsx") ############

data.head()
```

```python
# 논문명, 키워드, 국문초록 없는 값 삭제
data = data[data['비고'].isnull()]
data.head()
```

```python
data.shape
```

```python
filtered_data = data.drop(columns=['NO', '발행년', '권(호)', '논문ID', '비고'])
filtered_data.head()
```

```python
# 논문명 + 키워드 + 국문초록
final_data = data.drop(columns=['권(호)', '논문ID', '논문명', '키워드(한국어)', '국문초록', '비고'])
final_data.columns = ['no', 'year']
final_data['kor_full'] = filtered_data['논문명'] + " " + filtered_data['키워드(한국어)'] + " " + filtered_data['국문초록']

final_data.head()
```

```python
final_data.shape
```

```python
final_data.to_csv('./kor_full.csv', index=False)
```
